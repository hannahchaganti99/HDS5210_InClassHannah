```python
import pandas as pd
import pandasql as ps
```

```python
data = pd.read_csv('complications.csv')
```

```python
data.head()
```

```python
sql = """
select city, count(distinct [provider_id]) as providers
from data
group by city
order by providers desc
"""
```

```python
cities = ps.sqldf(sql, locals())
```

```python
cities
```
