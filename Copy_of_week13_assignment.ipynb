### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
import pandas as pd

# Load the dataset
file_path = 'Patients_Data.csv'  # Replace with the actual path to your file
data = pd.read_csv(file_path)

# Get the number of rows and columns
rows, columns = data.shape

# Print the result
print(f"The dataset has {rows} rows and {columns} columns.")
```

### 2. Data Diversity

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
import pandas as pd

# Load your dataset
file_path = 'Patients_Data.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# Identify columns where all the values are the same
constant_columns = data.loc[:, data.nunique() == 1].columns.tolist()

# Print the result
if constant_columns:
    print("Columns with the same value for all rows:", constant_columns)
else:
    print("No columns have the same value for all rows.")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt

# Plotting the histogram for the Age column
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=15, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Patients', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd  # Assuming you have data loaded in 'data' from previous cells

# Assuming 'data' contains your DataFrame
new_data = data.copy()  # Creating a copy of the original DataFrame as 'new_data'

# Age and Gender combination
plt.figure(figsize=(10, 6))
for gender in new_data['Gender'].unique():
    subset = new_data[new_data['Gender'] == gender]

plt.hist(subset['Age'], bins=15, alpha=0.6, label=gender)  # Plotting within the loop

plt.title('Age Distribution by Gender', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'new_data' or 'data' contains the data you want to visualize
# If you have a separate DataFrame called 'updated_data', replace 'new_data' with it
updated_data = new_data  # Or updated_data = data

# Boxplot: Age by Ethnicity and Gender
plt.figure(figsize=(12, 6))
sns.boxplot(x='Ethnicity', y='Age', hue='Gender', data=updated_data, palette='Set2')
plt.title('Age Distribution by Ethnicity and Gender', fontsize=14)
plt.xlabel('Ethnicity', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
